```python
import sphn
import json
from pathlib import Path

paths = [str(f) for f in Path(
    "/users/s1f102201582/projects/mhcc-moshi/moshi/data/v2/data_stereo"
).glob("*.wav")]
durations = sphn.durations(paths)

with open("/users/s1f102201582/projects/mhcc-moshi/moshi/data/v2/data.jsonl", "w") as fobj:
    for p, d in zip(paths, durations):
        if d is None:
            continue
        json.dump({"path": p, "duration": d}, fobj)
        fobj.write("\n")
```
