```python
import sentencepiece
from functools import reduce
```

```python
def tokenize(
    tokenizer: sentencepiece.SentencePieceProcessor,
    text: str,
    bos: bool = True,
    alpha: float | None = None,
):
    """Tokenize the given string, accounting for new lines, potentially adding a BOS token."""
    if alpha is not None:
        tokens = tokenizer.encode(
            text.split("\n"), enable_sampling=True, alpha=alpha, nbest_size=-1
        )
    else:
        tokens = tokenizer.encode(text.split("\n"))
        print(tokens)
    tokens = reduce(lambda a: [ *a ], tokens)
    if bos:
        tokens = [tokenizer.bos_id(), *tokens]
    return tokens
```

```python
tokenizer = sentencepiece.SentencePieceProcessor(model_file="./jmoshi_model/tokenizer_spm_32k_3.model")

text = "こんにちは、西方。調子はどう？"

print(tokenize(tokenizer, text))
```
